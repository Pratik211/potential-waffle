```python
# Conversion to datetime format and detection for variance present in data for calculating minutes difference. 

progress_df['started_at'] =  pd.to_datetime(progress_df['started_at'], format='%Y-%m-%d %H:%M:%S.%f')
progress_df['finished_at'] =  pd.to_datetime(progress_df['finished_at'], format='%Y-%m-%d %H:%M:%S.%f')
progress_df['mins_spent'] = progress_df['finished_at'] - progress_df['started_at']
progress_df['mins_spent'] = progress_df['mins_spent'].astype('timedelta64[m]')
progress_df['mins_diff'] = progress_df['duration_minutes'] - progress_df['mins_spent']
```

# Minutes Difference with Minutes spent is more on some cases which can be a case of deviation in Learning time

```python
def get_count_map(data, token):
    return list({val : data[token].value_counts()[val] for val in data[token].unique().tolist()}.items())

def merge_multiple_dfs(dfs, joinkey):
    if len(dfs) < 2:
        raise valueError
    merged_df = pd.merge(dfs.pop(), dfs.pop(), how='inner', on=joinkey)
    if len(dfs) != 0:
        for df in dfs:
            merged_df = pd.merge(merged_df, df, how='inner', on=joinkey)
    return merged_df
        
```

# Stats based on Different Idkey and Relation

```python
# Creation of two dataframe based on a suspicion one and sincere for cheating case based on unique user ID.

suspicous_df = progress_df[ (progress_df['mins_spent'] != progress_df['duration_minutes']) & 
                                   (progress_df['mins_spent'] == progress_df['mins_spent'])]
sincere_df = progress_df[ (progress_df['mins_spent'] == progress_df['duration_minutes']) & 
                                   (progress_df['mins_spent'] == progress_df['mins_spent'])]
```

```python
idkey = 'user_id'
user_stats_df = merge_multiple_dfs([
                                    pd.DataFrame.from_records(get_count_map(suspicous_df, idkey), 
                                                   columns=[idkey, 'cheated_count']), 
                                    pd.DataFrame.from_records(get_count_map(sincere_df, idkey), 
                                                   columns=[idkey, 'completion_count']),
                                    pd.DataFrame.from_records(get_count_map(progress_df, idkey), 
                                                   columns=[idkey, 'materials_attempted']),
                                    pd.DataFrame.from_records(get_count_map(user_df[user_df['is_mastered']], idkey), 
                                                   columns=[idkey, 'mastered_skill_count']),
                                    pd.DataFrame.from_records(get_count_map(user_df, idkey), 
                                                   columns=[idkey, 'skill_count'])
                                   ], joinkey=idkey)
```

# User class

```python
from pprint import pprint

class user:
    def __init__(self, user_id: str, skills_dict: dict, stats_info: dict):
        self.user_id = user_id
        self.skills_dict = skills_dict
        self.stats_info = stats_info
        
    def describe(self):
        print(f'ID: {self.user_id}')
        print('\nSkills:')
        pprint(self.skills_dict)
        print('\nStats:')
        pprint(self.stats_info)

def build_userdb():
    userdb = []
    user_stats_cols = list(user_stats_df.columns)
    user_stats_cols.remove('user_id')
    for user_id in user_stats_df['user_id']:
        skills_dict = dict(zip(user_df[user_df['user_id'] == user_id].skill_id, user_df[user_df['user_id'] == user_id].is_mastered))
        stats_info = user_stats_df[user_stats_df['user_id'] == user_id][user_stats_cols].to_dict('records')
        stats_info = stats_info[0]  if isinstance(stats_info, list) else stats_info
        userdb.append(user(user_id, skills_dict, stats_info))
    return userdb

userlist = build_userdb()
```
